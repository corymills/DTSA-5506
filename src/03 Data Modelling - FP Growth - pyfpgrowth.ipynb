```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth, association_rules
import pyfpgrowth
import pickle
```

```python
df = pd.read_pickle("../data/processed.pkl")
```

```python
with open("../data/data_element_details.pkl", "rb") as file:
    data_element_details = pickle.load(file)
```

```python
df.columns
```

```python
drop_cols = []
for value in data_element_details.values():
    if value['Section Name'] not in ["Health Care Access", "Chronic Health Conditions"] and value['Label'] in df.columns:
        drop_cols.append(value['Label'])
df = df.drop(drop_cols, axis=1)
```

```python
transactions = []
for _, row in df.iterrows():
    transaction = [f"{col}:{val}" for col, val in zip(df.columns[:20], row[:20]) if pd.notna(val)]
    transactions.append(transaction)
```

```python
transactions = transactions[:100]
transactions
```

```python
patterns = pyfpgrowth.find_frequent_patterns(transactions, .3)

# print("Frequent Patterns:")
# for pattern, support in patterns.items():
#     print(f"{pattern}: {support}")
```

```python
# Generate association rules
min_confidence = 0.7  # Minimum confidence threshold
rules = pyfpgrowth.generate_association_rules(patterns, min_confidence)

print("\nAssociation Rules:")
for rule in rules:
    antecedent = rule[0]
    consequent = rule[1]
    confidence = rule[2]
    print(f"{antecedent} -> {consequent} (Confidence: {confidence})")
```
